#Exercícios

```python
'''
1- Faça um programa que leia dois números
inteiros e exiba a soma deles.
'''
n1 = int(input("Digite o primeiro número: "))
n2 = int(input("Digite o segundo número: "))
print(f"A soma dos dois números corresponde a {n1 + n2}")
```

```python
'''
2- Faça um programa que leia a nota que um aluno tirou em P1(Prova 1) e P2(Prova 2).
Calcule a média aritmética. Exiba esse resultado na tela.
'''
P1 = float(input("Digite a nota da prova 1: "))
P2 = float(input("Digite a nota da prova 2: "))
media = (P1 + P2) / 2
print(f"A média do aluno é {media}")
```

```python
'''
3- Leia um nome armazenado em uma variável e imprima:
Olá, nome_pessoa! (nome_pessoa é a variável)
'''
nome = str(input("Digite seu nome: "))
print(f"Olá, {nome}!")
```

```python
'''
4- Solicite que o usuário digite seu nome.
Solicite que o usuários digite seu ano de nascimento (formato AAAA).
Exiba na tela: nome_usuario nasceu em ano_nascimento!
(nome_usuario e ano_nascimento são as variáveis)
'''
nome = str(input("Digite seu nome: "))
ano_nascimento = int(input("Digite seu ano de nascimento (AAAA): ")) # Added format hint
print(f"{nome} nasceu em {ano_nascimento}!")
```

```python
'''
5- Solicite que o usuário digite seu nome e sua idade.
Imprima “Olá, nome! Você tem idade anos?”
Imprima o ano de nascimento da pessoa (ignore meses e dias)
Imprima o nome da pessoa idade vezes (João tem 18, imprime JoãoJoão...)
'''
nome = str(input("Digite seu nome: "))
idade = int(input("Digite sua idade: "))
print(f"Olá, {nome}! Você tem {idade} anos?")
print(f"{nome} nasceu em {2025 - idade}!")
print((nome) * idade)
```

```python
'''
6- Solicite ao usuário que digite três notas
depois calcule e exiba a média ponderada.
'''
n1 = float(input("Digite a primeira nota: "))
n2 = float(input("Digite a segunda nota: "))
n3 = float(input("Digite a terceira nota: "))
media = (n1 * 2 + n2 * 3 + n3 * 5) / 10
print(f"A média ponderada é {media}")
```

```python
'''
7- Solicite ao usuário três notas e seus
respectivos pesos, depois calcule e exiba a
média ponderada.
'''
n1 = float(input("Digite a primeira nota: "))
n2 = float(input("Digite a segunda nota: "))
n3 = float(input("Digite a terceira nota: "))
peso1 = float(input("Digite o peso da primeira nota: "))
peso2 = float(input("Digite o peso da segunda nota: "))
peso3 = float(input("Digite o peso da terceira nota: "))
media = (n1 * peso1 + n2 * peso2 + n3 * peso3) / (peso1 + peso2 + peso3)
print(f"A média ponderada é {media}")
```

```python
'''
8- Peça ao usuário um número e exiba a soma
de seus dígitos.
'''
numero = int(input("Digite um número: "))
soma = 0
while numero > 0:
    digito = numero % 10
    soma += digito
    numero //= 10
print(f"A soma dos dígitos é {soma}")
```
